```python
import pandas as pd
```

```python
metadata_df = pd.read_csv("../data/metadata/metadata_book.csv")
```

```python
able_download_df = metadata_df[~metadata_df["text_download_url"].isna()]
```

```python
able_download_df.head()
```

```python
unknown_name_df = able_download_df[able_download_df["name"] == "unknown"]
unknown_name_df.head()
```

```python
def get_name_from_url(url):
  return url.split("/")[-1].split(".")[0]


unknown_name_df["name"] = unknown_name_df["audio_url"].apply(get_name_from_url)
unknown_name_df
```

```python
def get_name_from_url(url):
  return url.split("/")[-1].split(".")[0]


able_download_df.loc[:, "name"] = able_download_df["audio_url"].apply(get_name_from_url)
```

```python
able_download_df.to_csv("../data/metadata/able_download_metadata.csv", index=False)
```
