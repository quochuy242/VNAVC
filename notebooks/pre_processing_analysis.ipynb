```python
import pandas as pd
from pathlib import Path
```

```python
audio_path = Path("../data/audio/processed")
text_path = Path("../data/text/processed")

len(list(audio_path.glob("*.wav"))), len(list(text_path.glob("*.txt")))
```

```python
def pairing(audio_path, text_path):
  audio_files = list(audio_path.glob("*.wav"))
  text_files = list(text_path.glob("*.txt"))

  audio_files.sort()
  text_files.sort()

  paired_files = []
  i_text, i_audio = 0, 0
  while i_text < len(text_files) and i_audio < len(audio_files):
    text_file = text_files[i_text]
    audio_file = audio_files[i_audio]

    if text_file.stem == audio_file.stem:
      paired_files.append((audio_file.stem, audio_file, text_file))
      i_text += 1
      i_audio += 1
    elif text_file.stem < audio_file.stem:
      i_text += 1
    else:
      i_audio += 1

  return paired_files
```

```python
pairs = pairing(audio_path, text_path)
pair_df = pd.DataFrame(pairs, columns=["name", "audio", "text"])
len(pair_df)
```
